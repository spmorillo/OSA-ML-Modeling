# Importing libraries

```python
from IPython.display import display
import time
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Importing speech for male dataset

```python
# Importing the df from excel file
file = 'OSA_propio_classification_speech.csv' #OSA_propio_classification.csv
df = pd.read_csv(file)
#df.set_index("Patient", inplace=True)
display(df)
print("\nStats based on OSA feature")
display(df.groupby('OSA').describe())
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler


#Selecting features for the model
X = df.drop(["IAH","OSA","Patient"],axis=1)
X = pd.get_dummies(X, drop_first=False,
                   columns=["Gender","Smoker","Snorer","Illness"])
y = df["OSA"]

scaler = StandardScaler()
Xnorm = scaler.fit_transform(X)
scaler_2 = MinMaxScaler()
Xnorm_2 = scaler_2.fit_transform(X)
```

# Filter: Heatmaps

```python
#Global
plt.figure(figsize=(12,12))
hm = sns.heatmap(df.corr(),cmap="RdYlGn", annot=False)
bottom, top = hm.get_ylim()
hm.set_ylim(bottom + 0.5, top - 0.5)
hm.set_xticklabels(hm.get_xmajorticklabels(), fontsize = 15)
hm.set_yticklabels(hm.get_ymajorticklabels(), fontsize = 15) 
plt.show()
```

```python
#Local
filtered_columns = list()
substring_matching = ["A_", "E_", "I_", "O_", "U_",
                      "Form", "Ancho",
                      "1", "2","3","4"]

for substring in substring_matching:
    filtered = list(filter(lambda x: substring in x, df.columns))
    filtered_columns.append(filtered)
    
for col in filtered_columns:
    plt.figure(figsize=(12,12))
    hm = sns.heatmap(df[col].corr(),cmap="RdYlGn", annot=False)
    bottom, top = hm.get_ylim()
    hm.set_ylim(bottom + 0.5, top - 0.5)
    hm.set_xticklabels(hm.get_xmajorticklabels(), fontsize = 13)
    hm.set_yticklabels(hm.get_ymajorticklabels(), fontsize = 13) 
    plt.show()
```

# Filter: Univariate Selection
https://towardsdatascience.com/feature-selection-techniques-in-machine-learning-with-python-f24e7da3f36e

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

#Applying SelectKBest to extract the 10 best features
bestfeatures = SelectKBest(score_func=chi2, k=10)
fit = bestfeatures.fit(Xnorm_2,y)

#Concatenating dataframes for better visualization
dfscores = pd.DataFrame(fit.scores_)
dfpvalues = pd.DataFrame(fit.pvalues_)
dfcolumns = pd.DataFrame(X.columns)
featureScores = pd.concat([dfcolumns,dfscores, dfpvalues],axis=1)
featureScores.columns = ['Feature','Score',"p-values"]

#Printing and plotting the best features
feature_sorted = featureScores.sort_values("Score",ascending=False)
display(feature_sorted.head(10))

for col in ["Score","p-values"]:
    plt.figure(figsize=(15,10))
    ax = sns.barplot(x='Feature', y=col, data=feature_sorted.iloc[:6], palette="Set2")
    ax.set_xticklabels(feature_sorted['Feature'],rotation=0, fontsize=20)
    ax.set_xlabel("Feature",fontsize=20)
    ax.set_ylabel(col,fontsize=20)
    plt.show
```

# Wrapper: Recursive Feature Elimination

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

rfe_selector = RFE(estimator=LogisticRegression(solver='liblinear'),
                   n_features_to_select=5,
                   step=10
                  )
rfe_selector.fit(Xnorm_2, y)

rfe_support = rfe_selector.get_support()
rfe_feature = X.loc[:,rfe_support].columns.tolist()
ranking = sorted(zip(rfe_selector.ranking_, X.columns))

print("Top features:")
for feature in ranking[:20]:
    print(feature)
```

```python
#RFE with CV
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedKFold
from sklearn.feature_selection import RFECV

#SVC IS REALLY SLOW!!!
# Creating the RFE object and computing a cross-validated score.
rfecv = RFECV(estimator=LogisticRegression(solver='liblinear'),
              step=1,
              cv=StratifiedKFold(10),
              scoring='accuracy',
              verbose=0,
             )
rfecv.fit(Xnorm_2, y)

rfe_support = rfecv.get_support()
rfe_feature = X.loc[:,rfe_support].columns.tolist()
ranking = sorted(zip(rfecv.ranking_, X.columns))

print(rfe_feature)

print("Optimal number of features: %d" % rfecv.n_features_)
print("Top features:")
for feature in ranking[:rfecv.n_features_]:
    print(feature)

plt.figure(figsize=(15,10))
sns.set()
ax = sns.lineplot(x=range(1, len(rfecv.grid_scores_) + 1), y=rfecv.grid_scores_, palette="Set2")
ax.set_xlabel('Number of Features',fontsize=20)
ax.set_ylabel('CV Score (Accuracy)',fontsize=20)

plt.show()
```

# Embedded: Feature Importance
https://towardsdatascience.com/the-5-feature-selection-algorithms-every-data-scientist-need-to-know-3a6b566efd2

```python
from sklearn.ensemble import ExtraTreesClassifier

#Using inbuilt class feature_importances of tree based classifiers
model = ExtraTreesClassifier()
model.fit(X,y)

#Creating dataset of feature importances for better visualization
dfimportance = pd.DataFrame(model.feature_importances_)
dfcolumns = pd.DataFrame(X.columns)
feat_importances = pd.concat([dfcolumns,dfimportance],axis=1)
feat_importances.columns = ['Feature','Importance']

#Displaying it
df_sort = feat_importances.sort_values("Importance",ascending=False)
display(df_sort.head(6))

plt.figure(figsize=(15,10))
ax = sns.barplot(x='Feature', y='Importance', data=df_sort.iloc[:6], palette="Set2")
ax.set_xticklabels(df_sort['Feature'],rotation=0, fontsize=20)
ax.set_xlabel('Feature',fontsize=20)
ax.set_ylabel('Importance',fontsize=20)
plt.show
```

# Embedded: Select from Model

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import LogisticRegression

embeded_lr_selector = SelectFromModel(LogisticRegression(penalty="l1"),
                                      max_features=5)
embeded_lr_selector.fit(Xnorm_2, y)

embeded_lr_support = embeded_lr_selector.get_support()
embeded_lr_feature = X.loc[:,embeded_lr_support].columns.tolist()
for feature in embeded_lr_feature:
    print(feature)
```

# Dimensionality Reduction: PCA
https://towardsdatascience.com/visualising-high-dimensional-datasets-using-pca-and-t-sne-in-python-8ef87e7915b

```python
from sklearn.decomposition import PCA

scaler = StandardScaler()
Xnorm = scaler.fit_transform(X)

pca = PCA(50) 
Xnorm_pca = pca.fit_transform(Xnorm)

plt.figure(figsize=(15,10))
sns.set()
ax = sns.lineplot(x=list(range(50)),
                  y=np.cumsum(pca.explained_variance_ratio_),
                  palette="Set2")
ax.set_xlabel('Number of Components',fontsize=30)
ax.set_ylabel('Cumulative Explained variance',fontsize=30)
plt.show()
```

```python
pca = PCA(3) 
Xnorm_pca = pca.fit_transform(Xnorm)

df['pca-one'] = Xnorm_pca[:,0]
df['pca-two'] = Xnorm_pca[:,1] 
df['pca-three'] = Xnorm_pca[:,2]

print('Number of components: ' + str(pca.n_components_))
print('Explained variation per principal component:\n' + str(pca.explained_variance_ratio_))

plt.figure(figsize=(16,10))
ax = sns.scatterplot(
    x="pca-one", y="pca-two",
    hue="OSA",
    palette="Set2",
    data=df,
    s=300
)
ax.set_xlabel("pca-one",fontsize=30)
ax.set_ylabel("pca-two",fontsize=30)
plt.setp(ax.get_legend().get_texts(), fontsize='20') # for legend text
plt.setp(ax.get_legend().get_title(), fontsize='30') # for legend title
plt.show()

import plotly.express as px

fig = px.scatter_3d(df,
                    x='pca-one',
                    y='pca-two',
                    z='pca-three',
                    color='OSA',
                    color_discrete_sequence=[
                        "rgb(0, 100, 0)",
                        "rgb(255, 140, 0)"
                    ],
                   )
fig.update_traces(marker=dict(size=3))
fig.show()
```

# Visualization: t-SNE

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, verbose=1, perplexity=40, n_iter=300)
tsne_results = tsne.fit_transform(Xnorm[:,:40])

df['tsne-2d-one'] = tsne_results[:,0]
df['tsne-2d-two'] = tsne_results[:,1]

plt.figure(figsize=(16,10))
ax = sns.scatterplot(
    x="tsne-2d-one", y="tsne-2d-two",
    hue="OSA",
    palette="Set2",
    data=df,
    s=300
)
ax.set_xlabel("tsne-one",fontsize=30)
ax.set_ylabel("tsne-two",fontsize=30)
plt.setp(ax.get_legend().get_texts(), fontsize='20') # for legend text
plt.setp(ax.get_legend().get_title(), fontsize='30') # for legend title
plt.show()


tsne = TSNE(n_components=3, verbose=1, perplexity=40, n_iter=300)
tsne_results = tsne.fit_transform(Xnorm_2[:,:40])

df['tsne-3d-one'] = tsne_results[:,0]
df['tsne-3d-two'] = tsne_results[:,1]
df['tsne-3d-three'] = tsne_results[:,2]

fig = px.scatter_3d(df,
                    x='tsne-3d-one',
                    y='tsne-3d-two',
                    z='tsne-3d-three',
                    color='OSA',
                    color_discrete_sequence=[
                        "rgb(0, 100, 0)",
                        "rgb(255, 140, 0)"
                    ],
                   )
fig.update_traces(marker=dict(size=3))
fig.show()
```
