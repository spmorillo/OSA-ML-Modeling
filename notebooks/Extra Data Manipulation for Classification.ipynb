```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
import numpy as np
import matplotlib.pyplot as plt
import missingno
%matplotlib inline
```

```python
# Importing the df from excel file
file = 'OSA_propio.csv'
df = pd.read_csv(file)
#df.set_index("Patient", inplace=True)
df
```

```python
#Creating feature for classification
def foo(x):
    if x<=10.0:
        return "Healthy"
    elif x>=30.0:
        return "Severe"

df = df.loc[(df["IAH"]<=10.0) | (df["IAH"]>=30.0)]
df["OSA"] = df["IAH"].apply(foo)
df = df.loc[df["Gender"]=="hombre"]    
df["OSA"].value_counts()
```

```python
# Read Excel with extreme male cases
file = 'Formants_Bandwidth_OSA_male_speakers.xlsx'

# Load spreadsheet
xl = pd.ExcelFile(file)

df_OSA_speech = xl.parse('SPEECH_FEATURES')

df_OSA_speech
```

```python
df_OSA_speech.describe().iloc[1:].round(2)
```

```python
df_OSA_speech = df_OSA_speech.rename(columns = {"PATIENT": "Patient"})
```

```python
df_OSA_inner = pd.merge(df, df_OSA_speech, on='Patient', how='inner')
```
